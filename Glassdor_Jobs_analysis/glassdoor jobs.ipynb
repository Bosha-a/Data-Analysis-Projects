```python
def add_skill_columns(df, column_name='Job Description'):
    df['python']  = df[column_name].apply(lambda x: 1 if "python" in x.lower() else 0)
    df['excel']   = df[column_name].apply(lambda x: 1 if "excel" in x.lower() else 0)
    df['hadoop']  = df[column_name].apply(lambda x: 1 if "hadoop" in x.lower() else 0)
    df['spark']   = df[column_name].apply(lambda x: 1 if "spark" in x.lower() else 0)
    df['aws']     = df[column_name].apply(lambda x: 1 if "aws" in x.lower() else 0)
    df['tableau'] = df[column_name].apply(lambda x: 1 if "tableau" in x.lower() else 0)
    df['big_data']= df[column_name].apply(lambda x: 1 if "big_data" in x.lower() else 0)
    return df
```

```python
df = add_skill_columns(df)
```
